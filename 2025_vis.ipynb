```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.histplot(df['FinishPos'], kde=True, color='skyblue')
plt.title('Distribution of Race Finish Positions')
plt.xlabel('Finish Position')
plt.ylabel('Frequency')
plt.show()
```

```python
avg_points_per_team = df.groupby('Team')['Points'].mean().sort_values(ascending=False)

plt.figure(figsize=(14, 7))
sns.barplot(x=avg_points_per_team.index, y=avg_points_per_team.values, palette='viridis')
plt.title('Average Points per Team')
plt.xlabel('Team')
plt.ylabel('Average Points')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
avg_points_per_team = df.groupby('Driver')['Points'].mean().sort_values(ascending=False)

plt.figure(figsize=(14, 7))
sns.barplot(x=avg_points_per_team.index, y=avg_points_per_team.values, hue=avg_points_per_team.index, palette='flare', legend=False)
plt.title('Average Points per Driver')
plt.xlabel('Driver')
plt.ylabel('Average Points')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns

if 'EventDate' in numerical_cols:
    numerical_cols = numerical_cols.drop('EventDate')

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='GridPos', y='FinishPos', data=df, alpha=0.6)
plt.title('Grid Position vs. Finish Position')
plt.xlabel('Grid Position')
plt.ylabel('Finish Position')
plt.show()
```

```python
status_counts = df['Status'].value_counts()

plt.figure(figsize=(12, 7))
sns.barplot(x=status_counts.index, y=status_counts.values, palette='plasma', hue=status_counts.index, legend=False)
plt.title('Distribution of Race Statuses')
plt.xlabel('Race Status')
plt.ylabel('Number of Occurrences')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
avg_position_gain = df.groupby('Driver')['PositionGain'].mean().sort_values(ascending=False)

plt.figure(figsize=(14, 7))
sns.barplot(x=avg_position_gain.index, y=avg_position_gain.values, palette='magma', hue=avg_position_gain.index, legend=False)
plt.title('Average Position Gain per Driver')
plt.xlabel('Driver')
plt.ylabel('Average Position Gain')
plt.xticks(rotation=45, ha='right')
plt.axhline(0, color='grey', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.heatmap(df.isnull(), cbar=True, cmap='magma', xticklabels=True, yticklabels=False)
plt.title('Enhanced Missing Values Heatmap with Magma Colormap and Labels')
plt.show()
```
